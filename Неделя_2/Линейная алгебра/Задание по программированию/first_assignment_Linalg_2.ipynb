```python
% matplotlib inline
from matplotlib import pylab as plt
import scipy.linalg as linalg
from numpy import sin , exp
def f(x):
    return sin(x/5.0)*exp(x/10.0) + 5*exp(-x/2.0)

polynom_rank = 1
vector_A1 = [
    [1 ** n for n in range(0 , polynom_rank + 1)],
    [15 ** n for n in range(0, polynom_rank + 1)]
]
vector_b1 = [f(1),f(15)]
solution = linalg.solve(vector_A1,vector_b1)
print solution
plt.plot(solution)
plt.show()

polynom_rank = 2
vector_A2 = [
    [1 ** n for n in range(0, polynom_rank + 1)],
    [8 ** n for n in range(0, polynom_rank +1)],
    [15 ** n for n in range(0,polynom_rank +1)]
]
vector_b2 = [f(1),f(8),f(15)]
solution_2 = linalg.solve(vector_A2, vector_b2)
print solution_2
plt.plot(solution_2)
plt.show()

polynom_rank = 3
vector_A3 = [
    [1 ** n for n in range(0, polynom_rank + 1)],
    [4 ** n for n in range(0, polynom_rank +1)],
    [10 ** n for n in range(0, polynom_rank +1)],
    [15 ** n for n in range(0,polynom_rank +1)]
]

vector_b3 = [f(1),f(4),f(10),f(15)]
solution_3 = linalg.solve(vector_A3, vector_b3)
print solution_3

plt.plot(solution_3)
plt.show()

```
