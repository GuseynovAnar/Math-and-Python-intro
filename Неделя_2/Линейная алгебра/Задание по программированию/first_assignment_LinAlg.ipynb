```python
import scipy.spatial.distance as distance
import numpy as np
import re
f = open('sentences.txt', 'r')
words = {}
line_index , word_index = 0,0

lines_count = sum(1 for _ in f)
f.seek(0)
for line in f:
    low = line.lower()
    spliting = re.split('[^a-z]', low)
    replaced_text = [x for x in spliting if x != '']
    
    for word in replaced_text:
        if word not in words:
            words[word] = {
                "index": word_index,
                "occureInLines": [0] * lines_count
            }
            word_index += 1
            
        elif words[word]["occureInLines"][line_index] != 0:
            continue
            
        words[word]["occureInLines"][line_index] = replaced_text.count(word)    
    line_index += 1
    
for element_value in (sorted (val for key,val in words.items())):
    for key,val in words.items():
        if val == element_value:
            print(key,val)
            
array = np.zeros((lines_count , len(words)))

for word in words:
    i, j = 0, words[word]["index"]
    for occ in words[word]["occureInLines"]:
        array[i, j] = occ
        i += 1
        
# np.savetxt("myArray.csv", array, delimiter=",") to check 

comparison = {}
a = array[0,] # vector A
line_index = 1 # начиная со 2 предложения , если считать отсчет начиная  с  1 
answer = open("submission_1.txt" , 'a')

for i in range(1, lines_count):
    b = array[i,] # vector B 
    dist = distance.cosine(a,b)
    comparison = {
        "distance" : dist,
        "num" : line_index
    }
    line_index += 1
    print comparison 
    answer.write("%s " %(dist))
answer.close()
```
